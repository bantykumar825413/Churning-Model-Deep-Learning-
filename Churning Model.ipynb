```python
import matplotlib.pyplot as plt 
import pandas as pd
```

```python
# Importing the dataset
```

```python
dataset=pd.read_csv('Churn_modelling.csv')
```

```python
X=dataset.iloc[:,3:13]
```

```python
y=dataset.iloc[:,13]
```

```python
X
```

```python
# Lets assign disticnt numerical value to objects datatype like Georgrahoy,Gender coulumn 
```

```python
geography=pd.get_dummies(X["Geography"],drop_first=True)
gender=pd.get_dummies(X["Gender"],drop_first=True)
```

```python
X=pd.concat([X,gender,geography],axis=1)
```

```python
X=X.drop(["Geography","Gender"],axis=1)
```

```python
X
```

```python
# Spliting the for train test
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
```

```python

# Feature Scaling the dataset
```

```python
from sklearn.preprocessing import StandardScaler
sc= StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
```

```python
# Moving to Artificial neoral network 
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LeakyReLU,PReLU,ELU
from keras.layers import Dropout
```

```python
# Initiallizing the ANN
```

```python
classifier= Sequential()
```

```python
classifier.add(Dense(units=6,kernel_initializer='he_uniform',activation='relu',input_dim=11))
```

```python
# adding second layer 
```

```python
classifier.add(Dense(units=6,kernel_initializer='he_uniform',activation='relu'))
```

```python
# output layer
```

```python
classifier.add(Dense(units=1,kernel_initializer='glorot_uniform',activation='sigmoid'))
```

```python
# Compiling the ANN
```

```python
classifier.compile(optimizer='Adamax',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
model_history=classifier.fit(X_train,y_train,validation_split=0.33,batch_size=10,epochs=100)
```

```python
# Listing the data in history
```

```python
print(model_history.history.keys())
```

```python
# visualizing the Data
```

```python
plt.plot(model_history.history['accuracy'],color='red')
plt.plot(model_history.history['val_accuracy'],color='blue')
plt.title('model_accuracy')
plt.grid()
plt.ylabel('Accuracy')
plt.xlabel("Epoch")
plt.legend(['Train','Test'],loc='upper left')
```

```python
# Predicting the X_train
```

```python
y_pred=classifier.predict(X_test)
y_pred=y_pred > 0.5
```

```python
# comparing the accuracy 
```

```python
from sklearn.metrics import accuracy_score
score=accuracy_score(y_pred,y_test)
```
