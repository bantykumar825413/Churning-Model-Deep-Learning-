```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('income.csv')
```

```python
df.head(3)
```

```python
# Preprocessig the Dataset
```

```python
df=df.drop(['fnlwgt','education','native.country','race'],axis=1)
```

```python
# Removing the balnk spce 
```

```python
df.isin(['?']).sum()
```

```python
df.replace('?',np.nan,inplace =True)
```

```python
df.isnull().sum()
```

```python
# lets drop this missing values (Rows)
```

```python
df.dropna(inplace=True)
```

```python
# LEst encode the targeted values
```

```python
df['income']=df['income'].map({
    '<=50K':0,
    '>50K':1
})
```

```python
# Lets check :
df['income'].value_counts()
```

```python
np.unique(df['income'])
```

```python
# Lets separate Dependent and independent feature
```

```python
X=df.drop('income',axis=1)
y=df['income']
```

```python
# ONE -Hot Encoding Categorical features
```

```python
X.select_dtypes(include='object').columns
```

```python
X=pd.get_dummies(X,drop_first=True)
```

```python
X
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=42,test_size=0.3)
```

```python
# Lets scale the data 
```

```python
X_train=scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
```

```python
# Lets build the ann

```

```python
import tensorflow as tf
from tensorflow import keras 
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
```

```python
model=Sequential()
```

```python
model.add(Dense(64,activation='relu',input_dim=X_train.shape[1]))
model.add(Dropout(0.3))

model.add(Dense(32,activation='relu'))
model.add(Dropout(0.3))

model.add(Dense(1,activation='sigmoid'))
```

```python
model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy'],
)
```

```python
y_train.unique()
```

```python
history=model.fit(X_train,y_train,epochs=50,batch_size=32,validation_split=0.3)
```

```python
y_pred=model.predict(X_test)
```

```python
y_pred=y_pred > 0.5
```

```python
from sklearn.metrics import accuracy_score
accuracy=accuracy_score(y_pred,y_test)
print(accuracy)
```
